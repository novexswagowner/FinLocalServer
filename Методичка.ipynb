## <font color="Teal"> Основные понятия и определения

1. **Сетевое программирование** — это область разработки программного обеспечения, фокусирующаяся на создании приложений, которые могут общаться между собой посредством сетевых протоколов. Эти приложения способны обмениваться данными и выполнением операций через различные виды сетей, включая интернет и внутренние корпоративные сети.

2. **TCP/IP (Transmission Control Protocol/Internet Protocol)** — это фундаментальный набор сетевых протоколов, который лежит в основе интернета. TCP обеспечивает надежную передачу данных между компьютерами в сети, гарантируя, что данные будут доставлены без ошибок и в правильном порядке, в то время как IP отвечает за маршрутизацию отправляемых пакетов данных к их конечному назначению.

3. **HTTP (Hypertext Transfer Protocol)** — это протокол прикладного уровня, используемый для передачи гипертекстовых документов, таких как веб-страницы. HTTP является основой данных для веба, позволяя пользователям получать веб-страницы и другие ресурсы через интернет.

4. **FTP (File Transfer Protocol)** — протокол, предназначенный для передачи файлов между компьютерами в сети. FTP позволяет пользователям загружать и скачивать файлы, а также управлять файлами и директориями на удаленном сервере.

5. **SMTP (Simple Mail Transfer Protocol)** — это стандартный протокол для передачи электронной почты через интернет. SMTP используется почтовыми серверами для отправки, приема и пересылки электронных сообщений между отправителями и получателями.

6. **Сокеты** — это конечные точки для отправки и приема данных в сети. В контексте сетевого программирования сокет представляет собой программный интерфейс для обмена данными между различными процессами через сетевые протоколы, как локальные, так и через интернет.

7. **UDP (User Datagram Protocol)** — протокол транспортного уровня, который используется для передачи данных без установления соединения, не гарантируя доставку, проверку ошибок или сохранение последовательности данных. UDP часто применяется в приложениях, требующих высокой скорости передачи с минимальной задержкой.

8. **Многопоточность** — это способность программы одновременно выполнять несколько задач, распределяя их по различным потокам выполнения. В сетевом программировании многопоточность позволяет обрабатывать несколько соединений одновременно, повышая производительность и отзывчивость приложений.

9. **Асинхронное программирование** — метод организации кода, позволяющий программе продолжать выполнение, не ожидая завершения длительных операций ввода-вывода. Это улучшает производительность приложений, особенно тех, которые зависят от сетевых запросов и обмена данными.

10. **Шифрование** — это процесс кодирования информации таким образом, что только авторизованные стороны могут ее прочитать. Шифрование играет критическую роль в защите конфиденциальности и целостности данных в сетевом программировании.

11. **Асимметричное шифрование** — метод шифрования, использующий пару ключей: один для шифрования (публичный ключ) и другой для дешифрования (приватный ключ). Это обеспечивает безопасный обмен данными, так как только обладатель приватного ключа может расшифровать сообщение.

12. **Симметричное шифрование** — метод, при котором один и тот же ключ используется как для шифрования, так и для дешифрования сообщений. Этот метод обеспечивает быстрое шифрование больших объемов данных, но требует безопасного способа обмена ключом между сторонами.

13. **Модель OSI (Open Systems Interconnection)** — это концептуальная модель, разработанная для стандартизации функций телекоммуникационной или компьютерной сети на семь уровней, от физического до прикладного. Модель помогает разработчикам и инженерам понимать и стандартизировать процессы передачи данных.

14. **GIL (Global Interpreter Lock)** — механизм в интерпретаторе Python, предотвращающий одновременное выполнение нескольких потоков Python-кода. GIL обеспечивает, что в любой момент времени активен только один поток, что важно для управления доступом к объектам памяти и предотвращения состояний гонки.

15. **Селекторы (Selectors)** — компоненты в Python, облегчающие эффективное управление множеством сетевых соединений. Они позволяют отслеживать события ввода-вывода на нескольких сокетах одновременно, обеспечивая эффективное мультиплексирование соединений без необходимости создания отдельных потоков для каждого соединения.

16. **Неблокирующий ввод/вывод (Non-blocking I/O)** — подход к сетевому программированию, при котором операции ввода-вывода не блокируют выполнение программы. Это позволяет приложению продолжать обработку других задач, пока ожидается завершение операции ввода-вывода, улучшая общую производительность и отзывчивость.

17. **Модель клиент-сервер** — архитектурная модель, в которой сервер предоставляет ресурсы или услуги, а клиенты инициируют запросы к серверу для их использования. Эта модель является основой для большинства сетевых приложений и сервисов.

18. **Протокол** — соглашение или стандарт, который определяет формат и последовательность сообщений между двумя или более устройствами или приложениями для выполнения задачи или обмена данными.

19. **Порт** — числовой идентификатор в компьютерной сети, используемый вместе с IP-адресом для идентификации конкретного процесса или типа службы на хосте. Порты позволяют одному хосту поддерживать множество параллельных соединений для различных служб.

20. **IP-адрес** — уникальный числовой идентификатор, присваиваемый каждому устройству, подключенному к сети. IP-адрес использу ется для идентификации устройства и обеспечения его доступности в сети.

21. **Хэш-функция** — функция, преобразующая входные данные в строку фиксированной длины, обычно используемая для проверки целостности данных. Хэш-функции играют важную роль в криптографии и безопасности данных.

22. **Функция выработки ключа на основе хэша (Key Derivation Function, KDF)** — метод, используемый для создания одного или нескольких криптографических ключей из общего секрета. KDF обеспечивает безопасное преобразование исходного секрета в один или несколько ключей для шифрования или аутентификации.

23. **RESTful API (Representational State Transfer)** — архитектурный стиль разработки веб-сервисов, который использует стандартные методы HTTP для обмена данными между клиентом и сервером. RESTful API делает акцент на простоте и универсальности интерфейсов.

24. **JSON (JavaScript Object Notation)** — текстовый формат обмена данными, основанный на JavaScript, но используемый многими языками программирования. JSON легко читаем человеком и легко анализируем программами, что делает его идеальным для обмена данными между сервером и клиентом.

## <font color="Teal"> Упражнение 1. Основы работы с сокетами в Python

**Цели упражнения:**

* Познакомиться с базовыми принципами сетевого программирования в Python.

* Научиться создавать сервер и клиент, использующие TCP-соединения для обмена сообщениями.

Сокеты — это конечные точки в сетевом соединении. Сокеты представляют собой абстрактный интерфейс для взаимодействия программ с сетевым стеком, позволяя отправлять и получать данные через сеть.

С помощью сокетов программы могут обмениваться данными независимо от локальных или глобальных сетей, используя различные протоколы транспортного уровня, такие как TCP и UDP. Это основа для создания сетевых приложений, таких как веб-серверы, чаты, системы передачи файлов и многие другие.

Сокеты используются в большинстве сетевых приложений. Интернет и локальные сети полны примерами использования сокетов, от простого обмена сообщениями между двумя компьютерами до сложных распределенных систем и веб-сервисов.

Понимание работы сокетов критически важно для разработки сетевых приложений. Знание основ сетевого программирования позволяет создавать эффективные и надежные приложения, которые могут общаться через сеть, обеспечивая обмен данными, синхронизацию и многое другое.

**Создание сервера:**

```python
import socket

def start_server():
    server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    server_socket.bind(('localhost', 9090))
    server_socket.listen()
    print("Сервер запущен и ожидает подключений...")
    client_socket, client_address = server_socket.accept()

    try:
        while True:
            data = client_socket.recv(1024)
            if not data:
                break
            client_socket.sendall(data)
    finally:
        client_socket.close()
        server_socket.close()

start_server()
```

В процессе создания сетевого взаимодействия между двумя программами через использование сокетов в Python, мы начинаем с основного шага: создания сокета. Вызов функции `socket.socket(socket.AF_INET, socket.SOCK_STREAM)` является начальной точкой, где `socket.AF_INET` указывает на использование сетевого протокола IPv4, а `socket.SOCK_STREAM` обозначает использование протокола TCP. Выбор именно этих параметров не случаен. TCP, будучи протоколом управления передачей, обеспечивает надежное соединение между отправителем и получателем. Он гарантирует, что все пакеты данных будут доставлены в том порядке, в котором они были отправлены, и подтверждает получение каждого пакета, что критически важно для многих приложений, требующих точности и надежности в обмене данными.

Далее, процесс `bind(('localhost', 9090))` привязывает созданный сокет к конкретному адресу и порту на локальной машине (`localhost`), обозначенному номером `9090`. Этот шаг критически важен, поскольку он определяет точку, к которой будут направляться входящие подключения. Без выполнения этого шага сокет остается анонимным и не может принимать входящие соединения. Указание `localhost` означает, что сервер будет принимать соединения, инициированные на той же машине, что и сам сервер.

Следующий этап, `listen()`, переводит сокет в режим ожидания входящих подключений. Это состояние, в котором сервер становится доступным для клиентов и готов к установлению новых соединений. Фактически, сервер в этот момент "слушает" сеть в ожидании новых запросов на подключение.

Когда клиент пытается подключиться к серверу, `accept()` играет ключевую роль, блокируя выполнение программы до момента установления соединения. После успешного подключения клиента, `accept()` возвращает новый сокет, который используется для коммуникации с клиентом, и адрес клиента. Этот механизм обеспечивает возможность непосредственного обмена данными между сервером и клиентом.

В ходе сессии связи сервер входит в бесконечный цикл, в котором с помощью метода `recv(1024)` принимает данные от клиента. Этот метод читает данные из сокета, ожидая получения не более 1024 байт за один вызов. Полученные данные затем отправляются обратно клиенту с использованием `sendall(data)`, что делает сервер эхо-сервером, отражающим все полученные сообщения обратно отправителю. Этот простой механизм обмена данными демонстрирует основные принципы работы с сокетами: чтение и отправку данных через сеть.

Завершающий этап включает в себя закрытие соединения после того, как клиент решает прекратить обмен данными, закрывая свой сокет. В ответ сервер также закрывает соответствующие сокеты, освобождая системные ресурсы и завершая сессию связи. Этот шаг подчеркивает важность корректного завершения сетевого взаимодействия, предотвращая утечки ресурсов и потенциальные проблемы с производительностью.

Таким образом, процесс создания, привязки, прослушивания, принятия подключений, обмена данными и закрытия сокетов образует основу сетевого программирования.

**Создание клиента:**

```python
import socket

def start_client():
    client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    client_socket.connect(('localhost', 9090))

    try:
        client_socket.sendall(b"Hello, Server!")
        response = client_socket.recv(1024)
        print(f"Ответ от сервера: {response.decode('utf-8')}")
    finally:
        client_socket.close()

start_client()
```

Когда речь заходит о создании клиентской части в сетевом взаимодействии, первый шаг аналогичен тому, что был предпринят на сервере: создание сокета. Этот процесс, выполненный с помощью вызова `socket.socket(socket.AF_INET, socket.SOCK_STREAM)`, служит для инициализации нового сокета, который работает с использованием протокола IPv4 (`AF_INET`) и TCP (`SOCK_STREAM`). Выбор TCP в качестве транспортного протокола объясняется его свойствами обеспечения надежности и последовательности доставки данных, что делает его предпочтительным вариантом для множества сценариев взаимодействия в сети.

После создания сокета следующим шагом является установление соединения с сервером. Для этого клиент использует метод `connect(('localhost', 9090))`, который инициирует попытку подключения к серверу, ожидающему входящие соединения на локальном адресе (`localhost`) и порту `9090`. Этот шаг критичен, поскольку без успешного установления соединения обмен данными между клиентом и сервером невозможен. Указание `localhost` означает, что клиент предполагает наличие сервера на том же устройстве; в более общих случаях может использоваться IP-адрес или доменное имя удаленного сервера.

После успешного подключения к серверу клиент переходит к обмену данными. Отправка сообщения серверу производится с помощью метода `sendall(b"Hello, Server!")`, где `b"Hello, Server!"` представляет собой байтовую строку, являющуюся содержимым сообщения. Выбор байтовой строки обусловлен низкоуровневой природой сетевого взаимодействия, в рамках которого данные передаются именно в бинарном формате. Метод `sendall` гарантирует, что все данные будут отправлены, при необходимости разбивая сообщение на несколько пакетов.

Получение ответа от сервера осуществляется через вызов `recv(1024)`, который читает данные из сокета. Аргумент `1024` указывает максимальный размер данных, который можно прочитать за один вызов функции. В зависимости от объема ответа сервера, может потребоваться несколько вызовов `recv` для полного приема сообщения. Полученные данные, будучи байтовой строкой, затем декодируются в текстовую строку (обычно в кодировке UTF-8) для удобства представления пользователю. ## <font color="Teal"> Упражнение 2. Многопоточное обслуживание клиентов

**Цели упражнения:**

* Освоить создание и управление потоками в Python для обработки нескольких клиентских соединений одновременно.

* Научиться применять многопоточность в сетевых приложениях.

Многопоточность в программировании — это мощная концепция, позволяющая одновременно выполнять несколько операций в рамках одного и того же процесса. Этот метод особенно важен в современных вычислительных задачах, где требуется эффективное использование ресурсов и уменьшение времени ожидания. В ядре многопоточности лежит понятие «потока выполнения» — базовой единицы, которая может обрабатывать задачу параллельно с другими потоками, тем самым повышая производительность и отзывчивость приложений.

В Python многопоточность реализуется с помощью модуля `threading`, который предлагает простые и интуитивно понятные инструменты для создания, запуска и управления потоками. Этот модуль позволяет разработчикам легко разделить основную логику программы на несколько параллельно выполняемых частей, облегчая таким образом реализацию задач, требующих асинхронной обработки или одновременного выполнения.

**Ключевые понятия многопоточности в Python включают:**

* **Поток (Thread)** — это основная единица выполнения в многопоточном программировании. Потоки работают параллельно в пределах одного процесса, каждый занимаясь своей задачей. Они делят ресурсы процесса (память и файловые дескрипторы), что делает потоки более легковесными по сравнению с процессами. Модуль `threading` в Python предоставляет класс `Thread` для создания и управления потоками.

* **GIL (Global Interpreter Lock)** — это механизм, используемый Python для синхронизации доступа к объектам Python между разными потоками. GIL предотвращает одновременное выполнение нескольких потоков в интерпретаторе Python, что означает, что в любой момент времени активен может быть только один поток. Хотя это может показаться ограничением, на практике GIL редко становится препятствием для операций ввода-вывода или сетевого общения, поскольку эти операции часто блокируются, освобождая GIL для выполнения других потоков.

* **threading.Thread:** — это класс в модуле `threading`, который используется для создания новых потоков. Чтобы определить задачу, выполняемую потоком, разработчики могут либо наследовать класс `Thread` и переопределить его метод `run`, либо указать целевую функцию при создании экземпляра `Thread`.

Многопоточность особенно полезна в сценариях, где приложение должно одновременно обрабатывать множество задач, таких как обслуживание нескольких сетевых соединений, выполнение задач ввода-вывода или ожидание данных от различных источников. Благодаря способности выполнять множество задач параллельно, многопоточные приложения могут значительно улучшить пользовательский опыт и эффективность обработки данных.

Понимание и правильное использование многопоточности открывает перед разработчиками Python широкие возможности для создания высокопроизводительных и масштабируемых приложений, способных эффективно управлять параллельными задачами и обеспечивать быстрое и надежное взаимодействие в современных вычислительных средах.

```python
import socket
import threading

def handle_client(client_socket):
    """Обрабатывает соединение с клиентом в отдельном потоке."""
    try:
        while True:
            message = client_socket.recv(1024)
            if not message:
                break  # Клиент закрыл соединение
            client_socket.sendall(message)  # Эхо-ответ
    finally:
        client_socket.close()

def start_server():
    """Запускает сервер и обрабатывает множество соединений."""
    server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    server_socket.bind(('localhost', 9090))
    server_socket.listen()
    print("Сервер запущен и ожидает подключений...")

    while True:
        client_sock, address = server_socket.accept()
        print(f"Принято соединение от {address}")
        client_thread = threading.Thread(target=handle_client, args=(client_sock,))
        client_thread.start()

start_server()
```

В рамках многопоточного эхо-сервера, когда сервер принимает новое соединение от клиента через вызов `accept()`, он сталкивается с необходимостью обработать это соединение, не прерывая возможность принять другие входящие соединения. Для решения этой задачи сервер использует мощный механизм многопоточности, предоставляемый модулем `threading` в Python. Сервер создает новый поток выполнения, используя класс `threading.Thread`, что позволяет ему обрабатывать каждое новое соединение в отдельном потоке, параллельно с основным потоком сервера и другими клиентскими потоками.

Этот новый поток инициализируется функцией `handle_client`, которой в качестве аргумента передается сокет подключенного клиента. Функция `handle_client` предназначена для чтения сообщений, получаемых от клиента, и отправки их обратно, действуя как зеркало (эхо). Это осуществляется в бесконечном цикле, в рамках которого функция ожидает данные от клиента. Как только данные поступают, они немедленно отправляются обратно тому же клиенту. Цикл продолжается до тех пор, пока соединение с клиентом активно; если клиент закрывает соединение, цикл прерывается, и функция `handle_client` завершает работу потока, закрывая сокет и освобождая ресурсы.

Для запуска потока и начала его работы используется метод `start()`, применяемый к экземпляру потока. Вызов `start()` не только инициирует выполнение функции, указанной при создании потока, но и делает это асинхронно, позволяя основному потоку сервера продолжить принимать другие входящие соединения без задержек. Таким образом, сервер обеспечивает параллельную обработку множества клиентских соединений, каждое в своем потоке, что значительно повышает его масштабируемость и способность обслуживать одновременно множество клиентов.

Этот подход демонстрирует мощь и гибкость многопоточности в контексте сетевого программирования, позволяя эффективно распределять ресурсы сервера и обеспечивать высокий уровень отзывчивости приложения. Применение многопоточности для обработки клиентских соединений является ключевой стратегией в разработке масштабируемых сетевых серверов, способных обслуживать высокую нагрузку и предоставлять качественные услуги большому количеству пользователей.

## <font color="Teal"> Упражнение 3. Реализация протокола Диффи-Хеллмана и асимметричное шифрование

**Цели упражнения:**

* Изучить и применить протокол Диффи-Хеллмана для обмена ключами в контексте асимметричного шифрования.

* Научиться реализовывать асимметричное шифрование для защиты данных в Python.

Протокол Диффи-Хеллмана является фундаментальным алгоритмом в области криптографии, который позволяет двум или более участникам создать общий секретный ключ, даже если они находятся в условиях, где их коммуникация может быть перехвачена или наблюдаема третьими лицами. Этот секретный ключ, согласованный между сторонами, затем может быть использован для симметричного шифрования, что позволяет защитить передаваемую информацию от несанкционированного доступа.

Основной принцип работы протокола основан на математической сложности задачи дискретного логарифмирования, делая вычисление исходных параметров из полученного общего ключа крайне затруднительным для потенциального атакующего.

Протокол использует два открытых параметра: большое простое число $p$ и базу $g$, которая является первообразным корнем по модулю $p$. Эти параметры не являются секретными и могут быть известны всем участникам коммуникации.

1. ***Выбор секретных чисел:*** Каждая сторона выбирает секретное число: $a$ для Алисы и $b$ для Боба, которые они сохраняют в тайне.

2. ***Генерация публичных ключей:*** Каждая сторона вычисляет свой публичный ключ и отправляет его другой стороне. Алиса вычисляет $A = g^a\bmod{p}$ и отправляет $A$ Бобу, а Боб вычисляет $B = g^b \bmod{p}$ и отправляет $B$ Алисе.

3. ***Вычисление общего секретного ключа:*** После получения публичного ключа от другой стороны, каждая сторона может вычислить общий секретный ключ. Алиса вычисляет $s = B^a\bmod{p}$, а Боб вычисляет $s = A^b\bmod{p}$. В результате, и Алиса, и Боб получают одинаковый общий секретный ключ $s$, который не может быть легко вычислен наблюдателем без знания секретных чисел $a$ или $b$.

Асимметричное шифрование или шифрование с открытым ключом использует пару ключей для каждого пользователя: публичный и приватный. Публичный ключ можно безопасно распространять, он используется для шифрования данных. Приватный ключ хранится в секрете и используется для дешифрования данных, зашифрованных соответствующим публичным ключом.

Асимметричное шифрование обеспечивает конфиденциальность, аутентификацию и целостность данных, но оно требует значительно больше вычислительных ресурсов по сравнению с симметричным шифрованием. Поэтому на практике часто используется гибридный подход: асимметричное шифрование применяется для безопасного обмена секретными ключами симметричного шифрования, которые затем используются для шифрования самих данных.

Протокол Диффи-Хеллмана играет ключевую роль в этом процессе, обеспечивая безопасный обмен секретными ключами симметричного шифрования через открытые каналы. Это позволяет сторонам эффективно защитить свою коммуникацию, несмотря на потенциальное присутствие угроз в сети.

**Генерация параметров Диффи-Хеллмана:**

```python
from cryptography.hazmat.backends import default_backend
from cryptography.hazmat.primitives.asymmetric import dh

parameters = dh.generate_parameters(generator=2, key_size=2048, backend=default_backend())
```

В начале процесса реализации протокола Диффи-Хеллмана стоит задача генерации ключевых параметров, которые лежат в основе всего процесса обмена ключами. Эти параметры включают в себя большое простое число $p$ и генератор $g$, который является первообразным корнем по модулю $p$. Выбор этих параметров критически важен, поскольку они определяют уровень безопасности и эффективность всего процесса шифрования.

Для генерации необходимых параметров в Python используется функция `dh.generate_parameters()`, предоставляемая криптографической библиотекой `cryptography`. Эта функция позволяет автоматически создать параметры, соответствующие требованиям безопасности и эффективности.

Аргументы функции `dh.generate_parameters()`:

* `generator=2` — выбор генератора $g$, равного 2, является общепринятым стандартом в протоколах Диффи-Хеллмана. Генератор $g$ используется для создания публичных ключей участников обмена. Важно, что $g$ является первообразным корнем по модулю $p$, что обеспечивает, что порождаемые им значения будут равномерно распределены по всей группе чисел от $1$ до $p-1$, максимизируя таким образом безопасность обмена ключами.

* `key_size=2048` — указание размера ключа в 2048 бит обеспечивает высокий уровень защиты от криптографических атак, основанных на возможности вычисления дискретного логарифма. Размер ключа является важным параметром безопасности, поскольку от него зависит сложность задачи для потенциального атакующего. Выбор ключа достаточной длины гарантирует, что даже при использовании мощных вычислительных ресурсов вычисление приватного ключа или общего секрета становится практически невозможным.

* `backend=default_backend()` — этот аргумент указывает на криптографическую библиотеку, используемую для выполнения операций. `default_backend()` представляет собой абстракцию, позволяющую `cryptography` автоматически выбирать наиболее подходящую реализацию для текущей системы, обеспечивая тем самым максимальную производительность и безопасность.

Генерация параметров Диффи-Хеллмана является первым и одним из самых важных шагов в установлении защищенного канала связи между сторонами. Правильный выбор параметров гарантирует, что основа для последующего обмена ключами и шифрования данных будет надежной и устойчивой к атакам.

После успешной генерации параметров Диффи-Хеллмана, следующим шагом в процессе установления защищенного канала связи является создание пары ключей для каждой из сторон, участвующих в обмене. Эти ключи играют центральную роль в обеспечении конфиденциальности и безопасности данных, передаваемых между сторонами.

Приватный ключ каждой стороны генерируется независимо и представляет собой случайно выбранное большое число, которое действует в качестве секрета, известного только владельцу ключа. В контексте протокола Диффи-Хеллмана, приватный ключ используется для вычисления публичного ключа и, в дальнейшем, для генерации общего секретного ключа, который будет использоваться для шифрования данных.

Процесс генерации приватного ключа начинается с использования сгенерированных ранее параметров Диффи-Хеллмана. Эти параметры обеспечивают математическую основу, необходимую для гарантии безопасности ключей, получаемых в результате.

После создания приватного ключа следующим шагом является генерация публичного ключа. Публичный ключ получается путем выполнения математических операций с использованием приватного ключа и общедоступных параметров Диффи-Хеллмана (большое простое число $p$ и генератор $g$). В частности, публичный ключ $A$ для Алисы или $B$ для Боба вычисляется как $g^a\bmod{p}$, где $a$ — приватный ключ стороны.

Этот процесс обеспечивает, что даже при знании публичного ключа и общедоступных параметров Диффи-Хеллмана, вычисление соответствующего приватного ключа становится вычислительно неразрешимой задачей без применения невероятных вычислительных ресурсов. Таким образом, публичный ключ может быть безопасно передан через незащищенные каналы связи другой стороне без риска раскрытия приватного ключа.

Приватный ключ является основой безопасности в асимметричном шифровании и в протоколе Диффи-Хеллмана. Он должен оставаться строго конфиденциальным, поскольку доступ к приватному ключу позволяет не только генерировать соответствующий публичный ключ, но и расшифровывать любые данные, зашифрованные с использованием этого публичного ключа. Поэтому сохранение приватного ключа в тайне от третьих лиц является критически важным для обеспечения безопасности обмена данными.

В результате, комбинация приватного и публичного ключей в протоколе Диффи-Хеллмана позволяет двум сторонам безопасно согласовать общий секретный ключ, даже находясь в условиях открытого канала связи, что делает этот протокол основой для многих современных систем защищенного обмена данными.

**Обмен публичными ключами:**

После того как участники протокола Диффи-Хеллмана сгенерировали свои публичные ключи, следующим критическим этапом является их обмен. Обмен публичными ключами позволяет каждой стороне использовать свой приватный ключ вместе с публичным ключом другой стороны для вычисления общего секретного ключа. Этот общий секретный ключ будет использоваться для зашифрования и расшифрования сообщений, обмениваемых между сторонами.

Обмен публичными ключами может происходить по различным каналам связи, в зависимости от доступной инфраструктуры и требований к безопасности. Важно отметить, что публичные ключи не являются секретными и их раскрытие не угрожает безопасности протокола, поэтому они могут передаваться открыто.

* В большинстве случаев обмен публичными ключами происходит через Интернет. Например, если речь идет о защищенном веб-соединении, сервер может отправить свой публичный ключ клиенту в составе цифрового сертификата при установлении соединения по протоколу HTTPS. Клиент, в свою очередь, может генерировать временный публичный ключ и отправлять его серверу для установления сеансового ключа.

* В сценариях, где требуется предварительное согласование, стороны могут обмениваться публичными ключами, отправляя их друг другу в виде файлов через электронную почту. Это распространенный метод в системах шифрованной электронной почты.

* Для упрощения обмена публичными ключами и повышения уровня доверия к ним, могут использоваться специализированные серверы публичных ключей. Эти серверы действуют как централизованные хранилища, где пользователи могут публиковать свои публичные ключи и извлекать публичные ключи других пользователей.

Примеры обмена публичными ключами

* Алиса и Боб хотят обмениваться зашифрованными сообщениями. Алиса генерирует пару ключей, отправляет свой публичный ключ Бобу через безопасный канал или сервер публичных ключей. Боб, получив публичный ключ Алисы, делает то же самое: генерирует пару ключей и отправляет свой публичный ключ Алисе. Теперь у Алисы есть публичный ключ Боба, и у Боба есть публичный ключ Алисы, что позволяет им обоим вычислить общий секретный ключ для шифрования и расшифрования сообщений.

* При посещении веб-сайта через HTTPS, браузер (клиент) запрашивает у сервера его публичный ключ в составе цифрового сертификата. После проверки подлинности сертификата и извлечения публичного ключа сервера, браузер использует его для шифрования и передачи сгенерированного симметричного ключа шифрования, который затем используется для зашифрованного обмена данными в течение сессии.

Эти примеры иллюстрируют, как процесс обмена публичными ключами лежит в основе создания безопасного канала связи между сторонами, позволяя им защищать конфиденциальность и целостность передаваемых данных.

**Вычисление общего секретного ключа:**

```python
from cryptography.hazmat.primitives import hashes
from cryptography.hazmat.primitives.kdf.hkdf import HKDF

shared_secret = private_key.exchange(other_public_key)
derived_key = HKDF(
    algorithm=hashes.SHA256(),
    length=32,
    salt=None,
    info=b'handshake data',
    backend=default_backend()
).derive(shared_secret)
```

По завершении обмена публичными ключами, каждая из сторон обладает всей необходимой информацией для генерации общего секретного ключа. Этот процесс включает в себя использование собственного приватного ключа и публичного ключа другой стороны. Полученный общий секретный ключ является уникальным результатом взаимодействия этих двух ключей, однако, в его "сыром" виде он еще не готов к использованию в криптографических операциях из-за переменной длины и потенциального наличия неструктурированных данных.

Для преобразования этого общего секретного ключа в ключ шифрования фиксированной длины, пригодный для последующего использования, применяется механизм HKDF (HMAC Key Derivation Function). HKDF состоит из двух основных этапов: извлечение и расширение. При этом, для начала работы с HKDF, необходимо определить ряд параметров:

* ***Алгоритм хэширования*** определяется как `hashes.SHA256()`, что указывает на использование криптографической хэш-функции SHA-256. Этот выбор обусловлен высокой стойкостью алгоритма SHA-256 к атакам, что делает его надежным выбором для создания безопасных ключей.

* ***Длина ключа***, задаваемая параметром `length=32`, указывает на то, что итоговый ключ шифрования будет иметь длину 32 байта (или 256 бит). Этот размер ключа считается достаточно большим для обеспечения высокого уровня безопасности при использовании современных алгоритмов шифрования.

* ***Соль (salt)***, в данном случае указанная как `None`, является опциональным параметром, который добавляет дополнительный уровень случайности к процессу выработки ключа. Отсутствие соли (`None`) в данном контексте указывает на прямое использование общего секретного ключа без дополнительных модификаций.

* ***Контекстные данные (info)***, заданные как `b'handshake data'`, позволяют ввести дополнительную информацию в процесс генерации ключа, тем самым уникализируя его для конкретного протокола или сессии обмена данными. Это может быть полезно для разграничения ключей, выработанных для различных целей или в разных сессиях общения.

* ***Бэкенд (backend)***, установленный в `default_backend()`, указывает на криптографическую библиотеку или интерфейс, который будет использоваться для выполнения криптографических операций. Выбор `default_backend()` означает автоматический выбор наиболее подходящего бэкенда, доступного в текущей системе.

После определения этих параметров HKDF процедура `derive(shared_secret)` применяется к общему секретному ключу для генерации итогового ключа шифрования фиксированной длины. Этот ключ теперь может быть использован обеими сторонами для безопасного шифрования и дешифрования сообщений, обеспечивая конфиденциальность и защиту данных в процессе их обмена.

**Использование секретного ключа для шифрования и дешифрования сообщений:**

После успешного завершения процесса обмена ключами и получения производного ключа через механизм HKDF, наступает этап шифрования данных, который обеспечивает конфиденциальность общения между сторонами. Для этого используется комбинация инициализирующего вектора (IV) и алгоритма AES в режиме CBC для создания зашифрованных сообщений.

```python
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
from cryptography.hazmat.backends import default_backend
from os import urandom

# Предположим, что derived_key — это ключ, полученный через HKDF
# Генерация случайного IV
iv = urandom(16)

# Создание объекта шифра AES с использованием производного ключа и режима CBC
cipher = Cipher(algorithms.AES(derived_key), modes.CBC(iv), backend=default_backend())

# Шифрование сообщения
encryptor = cipher.encryptor()
ciphertext = encryptor.update(b"Секретное сообщение") + encryptor.finalize()
```

Первым шагом в процессе шифрования является генерация инициализирующего вектора (IV). IV — это блок данных фиксированной длины, который используется вместе с секретным ключом для шифрования первого блока текста. Функция urandom из стандартной библиотеки Python генерирует случайный IV, обеспечивая уникальность каждой операции шифрования. Эта случайность критически важна для предотвращения повторения шифрованных блоков при шифровании одинаковых сообщений, повышая тем самым безопасность шифрованной передачи.

Следующим шагом является создание объекта шифра, который использует алгоритм AES (Advanced Encryption Standard) в режиме CBC (Cipher Block Chaining). AES — это широко применяемый и проверенный временем алгоритм симметричного шифрования, который был выбран в качестве стандарта правительством США и во всем мире благодаря его надежности и эффективности. Режим CBC дополнительно усиливает безопасность шифрования за счет включения предыдущего зашифрованного блока в процесс шифрования текущего блока. Это означает, что каждый блок шифруется не только с использованием секретного ключа, но и с учетом зашифрованных до него блоков, что делает невозможным расшифровку данных без знания всей цепочки.

Теперь, когда у нас есть зашифрованное сообщение (`ciphertext`) и IV, мы можем расшифровать сообщение, используя тот же ключ.

```python
# Дешифрование сообщения
decryptor = cipher.decryptor()
plaintext = decryptor.update(ciphertext) + decryptor.finalize()
```

После создания объекта шифра с использованием производного ключа и IV, сообщение может быть зашифровано. Метод update применяется к объекту шифра для передачи текста сообщения, который необходимо зашифровать. После передачи всех данных метод `finalize` вызывается для завершения процесса шифрования и гарантирует, что все данные были корректно обработаны. Результатом является зашифрованный текст, или `ciphertext`, который теперь можно безопасно передать по открытому каналу.

Этот процесс шифрования и последующего дешифрования с использованием AES в режиме CBC и сгенерированного секретного ключа позволяет сторонам обмениваться конфиденциальной информацией, сохраняя уверенность в том, что только они могут прочитать содержимое сообщений. Использование IV и режима CBC обеспечивает дополнительную безопасность, делая каждую операцию шифрования уникальной и защищенной от различных видов криптографических атак.

Процесс генерации ключа представляет собой критический этап, так как именно от надежности и уникальности ключа зависит безопасность всей системы шифрования. Сгенерированный таким образом ключ является секретным и должен быть известен только авторизованным сторонам, участвующим в обмене зашифрованными сообщениями.

После успешной генерации ключа следующим шагом является создание объекта шифра, который будет использоваться для шифрования и дешифрования данных. Это достигается путем передачи сгенерированного ключа в конструктор класса `Fernet`, результатом чего становится создание экземпляра `cipher`. Объект `cipher` включает в себя все необходимые алгоритмы и параметры для выполнения операций симметричного шифрования и дешифрования с использованием ключа, созданного на предыдущем шаге. Суть работы объекта `cipher` заключается в применении комплекса математических преобразований к исходным данным (для шифрования) или к зашифрованным данным (для дешифрования), основываясь на значении секретного ключа.

Эти два этапа — генерация ключа и создание объекта шифра — являются фундаментальными в процессе симметричного шифрования и обеспечивают базовую инфраструктуру для безопасного преобразования данных. Использование сгенерированного ключа как для шифрования, так и для дешифрования гарантирует, что только обладатели ключа могут получить доступ к содержанию зашифрованных сообщений, тем самым обеспечивая конфиденциальность и защиту информации от несанкционированного доступа.

Перед тем как перейти к шифрованию сообщения с использованием симметричного ключа и алгоритма Fernet, необходимо подготовить само сообщение к этому процессу. В компьютерной обработке данных текст обычно представлен в формате строки, который для целей криптографии требуется преобразовать в байты — основную единицу данных в компьютерных системах. Это преобразование выполняется с помощью метода `.encode()`, который трансформирует строку в последовательность байтов, используя указанную кодировку (по умолчанию UTF-8). Процесс кодирования критичен, так как алгоритмы шифрования, включая Fernet, оперируют именно байтами, а не текстом.

После преобразования сообщения в байты следует непосредственно процесс шифрования. Используя метод `cipher.encrypt(message)`, где `cipher` — это объект шифрования, созданный на основе секретного ключа, и `message` — подготовленное к шифрованию сообщение в байтовом представлении, мы активируем алгоритм Fernet. Этот метод применяет к байтам сообщения ряд криптографических преобразований, основанных на значении секретного ключа. В результате этих преобразований исходное сообщение становится неразличимым для кого бы то ни было, кроме обладателя секретного ключа, что и обеспечивает конфиденциальность информации. Возвращаемое значение метода `encrypt` — это зашифрованное сообщение, также представленное в виде байтовой строки. Эта строка может включать в себя не только зашифрованный текст, но и дополнительные метаданные, необходимые для дешифрования (например, соль или инициализирующий вектор), что делает процесс дешифрования возможным исключительно при наличии соответствующего секретного ключа.

Таким образом, подготовка сообщения к шифрованию и последующее его преобразование с помощью метода `encrypt` образуют основу процесса защиты информации с использованием симметричного шифрования. Эти шаги позволяют гарантировать, что конфиденциальные данные могут быть безопасно переданы или сохранены, будучи защищенными от несанкционированного доступа.

**Дешифрование сообщения:**

```python
# Дешифрование сообщения
decrypted_message = cipher.decrypt(encrypted_message)
print(f"Расшифрованное сообщение: {decrypted_message.decode()}")
```

Процесс дешифрования является обратным по отношению к шифрованию и представляет собой ключевую стадию в обеспечении доступности зашифрованной информации для её предполагаемого получателя. В контексте использования симметричного шифрования с помощью алгоритма Fernet, этот процесс начинается с применения метода `decrypt` к объекту шифрования `cipher`, который был инициализирован с тем же секретным ключом, что и для шифрования. Аргументом метода `decrypt` является зашифрованное сообщение `encrypted_message`, которое содержит байты данных, подвергшиеся криптографическим преобразованиям.

Когда метод `decrypt` вызывается, алгоритм Fernet использует секретный ключ для выполнения серии обратных криптографических операций над зашифрованными байтами. Эти операции эффективно «отменяют» шифрование, восстанавливая исходные байты сообщения. Важно отметить, что успешное дешифрование возможно только в случае использования точно такого же ключа, который был использован для шифрования данных. Это обеспечивает строгую связь между шифрованием и дешифрованием, гарантируя, что только участники, располагающие соответствующим секретным ключом, могут получить доступ к содержимому зашифрованных сообщений.

После того как исходные байты сообщения восстановлены, последним шагом в процессе дешифрования является их преобразование обратно в читаемую форму. Это достигается с помощью метода `decode`, который применяется к байтовой строке, возвращаемой методом `decrypt`. Метод `decode` трансформирует байты обратно в строку, используя заданную или по умолчанию кодировку (обычно UTF-8), тем самым делая расшифрованное сообщение доступным для чтения и дальнейшей обработки. Преобразование из байтов в строку является критически важным для восстановления исходного текста сообщения, так как в современных компьютерных системах текстовая информация обычно представлена именно в форме строк.

В совокупности, процессы дешифрования и последующего чтения сообщения позволяют получателю восстановить исходное содержимое зашифрованного сообщения, обеспечивая тем самым конфиденциальность обмена информацией между отправителем и получателем в условиях, когда доступ к секретному ключу строго контролируется и ограничен участниками взаимодействия.

## <font color="Teal"> Упражнение 5. Работа с UDP-сокетами

**Цели упражнения:**

* Ознакомиться с принципами работы протокола UDP в контексте сетевого программирования.

* Научиться создавать клиент и сервер, использующие UDP для обмена сообщениями.

При работе с протоколом UDP (User Datagram Protocol) важно понимать его основные характеристики и отличия от TCP (Transmission Control Protocol), другого ключевого протокола транспортного уровня. UDP является безсоединительным протоколом, что означает, что перед началом обмена данными между отправителем и получателем не устанавливается никакого специализированного соединения. Эта особенность делает UDP идеальным для сценариев, где критически важна скорость передачи за счет надежности, например, в приложениях реального времени (видеоконференции или многопользовательские онлайн-игры).

Одним из следствий безсоединительной природы UDP является то, что протокол не гарантирует доставку пакетов данных, их порядок и не предоставляет механизмов для контроля перегрузки сети. В отличие от TCP, который предусматривает подтверждение получения каждого пакета и повторную отправку в случае необходимости, UDP просто отправляет датаграммы и не требует подтверждения их получения. Это может привести к потере данных или их прибытию в непредсказуемом порядке, однако многие приложения компенсируют эти недостатки на более высоком уровне, реализуя собственные механизмы для обеспечения надежности и целостности передачи.

**Создание UDP-сервера:**

```python
import socket

def udp_server():
    server_socket = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
    server_socket.bind(('localhost', 9090))

    print("UDP сервер запущен и ожидает сообщений...")
    while True:
        message, client_address = server_socket.recvfrom(1024)
        print(f"Получено сообщение от {client_address}: {message.decode()}")
        server_socket.sendto(message, client_address)

udp_server()
```

При работе с UDP-сокетами процесс начинается с создания самого сокета, который является точкой входа для отправки и приема датаграмм в сети. Используя функцию `socket.socket()` из стандартной библиотеки Python `socket`, мы инициализируем сокет, указывая два ключевых параметра: `socket.AF_INET` и `socket.SOCK_DGRAM`. Параметр `socket.AF_INET` говорит, что наш сокет будет использовать сетевой интерфейс IPv4, самый распространенный протокол интернет-адресации. Параметр `socket.SOCK_DGRAM` указывает на то, что сокет предназначен для работы с протоколом UDP, который отправляет данные в виде датаграмм, не требуя установления постоянного соединения между отправителем и получателем.

После создания сокета следующим шагом является его привязка к определенному адресу и порту с помощью метода `bind()`. Этот шаг необходим для того, чтобы операционная система знала, какому приложению следует направлять входящие датаграммы, отправленные на указанный адрес и порт. В нашем случае сервер привязывается к локальному адресу (например, 'localhost') и выбранному порту (например, 9090), что делает его доступным для приема сообщений, отправленных на этот адрес.

Когда сервер готов к приему сообщений, он входит в бесконечный цикл, где ожидает входящие датаграммы от клиентов. Для этого используется метод `recvfrom()`, который блокирует выполнение программы до момента получения данных. Метод `recvfrom()` возвращает кортеж, содержащий полученные байты данных и адрес отправителя. Эта информация позволяет серверу не только обработать полученное сообщение, но и узнать, откуда оно пришло, что является основой для отправки ответа.

Наконец, для отправки эхо-ответа обратно клиенту сервер использует метод `sendto()`, который принимает два аргумента: данные для отправки и адрес получателя. В нашем случае сервер отправляет полученное сообщение назад к клиенту, используя адрес, полученный из метода `recvfrom()`. Это демонстрирует двунаправленную коммуникацию между клиентом и сервером с использованием UDP: клиент отправляет сообщение серверу, а сервер отвечает, отправляя сообщение обратно.

Весь этот процесс подчеркивает как основные особенности протокола UDP, так и методы работы с ним в Python. Создание и использование UDP-сокетов демонстрирует простоту и эффективность UDP для задач, где требуется минимальная задержка и не критична абсолютная надежность доставки каждого пакета данных.

**Создание UDP-клиента:**

```python
import socket

def udp_client():
    client_socket = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

    server_address = ('localhost', 9090)
    message = b"Привет, UDP сервер!"
    client_socket.sendto(message, server_address)

    response, _ = client_socket.recvfrom(1024)
    print(f"Ответ от сервера: {response.decode()}")

    client_socket.close()

udp_client()
```

Процесс создания и использования клиентского сокета предполагает несколько ключевых шагов, аналогичных тем, что были описаны для сервера, но с некоторыми особенностями, связанными с ролью клиента в обмене данными.

* Создание сокета

    На первом этапе клиентский код инициализирует UDP сокет с помощью функции `socket.socket()`, указывая `socket.AF_INET` для использования IPv4 адресации и `socket.SOCK_DGRAM` для указания на использование протокола UDP. Этот шаг создает сокет, который клиент будет использовать для отправки сообщений серверу и приема ответов от него. В отличие от сервера, клиентскому сокету обычно не требуется привязываться к конкретному порту с помощью метода `bind()`, так как порт для отправки и получения датаграмм может быть выбран операционной системой автоматически.

* Отправка сообщения

    Далее, клиент использует метод `sendto()` для отправки сообщения серверу. Этот метод принимает два аргумента: байты сообщения и кортеж, содержащий адрес и порт сервера. Сообщение предварительно кодируется в байты (если оно представлено в виде строки), чтобы соответствовать формату, с которым работает UDP. Метод `sendto()` позволяет направить сообщение непосредственно указанному получателю без установления долгосрочного соединения, демонстрируя безсоединительный характер протокола UDP.

* Получение ответа

    После отправки сообщения клиент ожидает ответ от сервера, используя тот же сокет и метод `recvfrom()`. Этот метод блокирует выполнение программы до тех пор, пока не будет получен ответ от сервера, возвращая данные ответа и адрес отправителя. Полученные данные обычно требуют декодирования из байтов обратно в строку (если ожидается текстовое сообщение), чтобы их можно было корректно интерпретировать и отобразить пользователю.

* Закрытие сокета

    Наконец, после обмена сообщениями с сервером, клиентский сокет закрывается с помощью метода `close()`. Этот шаг освобождает системные ресурсы, связанные с сокетом, и предотвращает утечки ресурсов, которые могут произойти при неоднократном создании и не закрытии сокетов. Закрытие сокета является важной частью управления сетевыми соединениями, обеспечивая корректное завершение сетевого взаимодействия и поддержание здоровья и стабильности как клиентского приложения, так и системы в целом.

Эти шаги иллюстрируют процесс создания, использования и закрытия UDP сокета на стороне клиента, демонстрируя простоту и эффективность протокола UDP для сценариев, где требуется быстрая и необременительная передача данных без гарантии доставки.

## <font color="Teal"> Упражнение 6. Использование селекторов для обработки множественных соединений

**Цели упражнения:**

Селекторы в Python предлагают элегантное решение для обработки множественных сетевых соединений в одном потоке исполнения, позволяя эффективно масштабировать сетевые приложения без необходимости прибегать к многопоточности или многопроцессорности. Этот механизм особенно ценен в контексте серверных приложений, где одновременно обрабатываются соединения от большого числа клиентов.

Ключевые аспекты работы с селекторами:

* ***Мультиплексирование ввода/вывода.*** Селекторы позволяют приложению отслеживать и реагировать на события ввода/вывода (например, готовность сокета к приему новых данных) на множестве сокетов одновременно. Это достигается путем регистрации сокетов в селекторе с указанием типа интересующих событий и колбэка, который должен быть вызван при их наступлении.

* ***Асинхронность.*** Благодаря неблокирующему режиму работы сокетов и использованию селекторов, приложение может асинхронно обрабатывать события ввода/вывода, что значительно увеличивает его отзывчивость и эффективность, избегая простоя в ожидании действий от одного соединения перед переходом к следующему.

* ***Управление событиями.*** Селекторы обеспечивают централизованное управление событиями ввода/вывода, автоматически вызывая зарегистрированные функции-обработчики (колбэки) для сокетов, на которых произошли зарегистрированные события. Это делает код приложения более чистым и понятным, сосредотачивая логику обработки соединений в одном месте.

**Создание мультиплексированного TCP-сервера с использованием селекторов:**

```python
import selectors
import socket

selector = selectors.DefaultSelector()

def accept_connection(server_socket):
    client_socket, client_address = server_socket.accept()
    print(f"Подключение от {client_address}")
    selector.register(client_socket, selectors.EVENT_READ, send_echo)

def send_echo(client_socket):
    data = client_socket.recv(1024)
    if data:
        client_socket.sendall(data)
    else:
        selector.unregister(client_socket)
        client_socket.close()

def start_server():
    server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    server_socket.bind(('localhost', 9090))
    server_socket.listen()
    server_socket.setblocking(False)
    selector.register(server_socket, selectors.EVENT_READ, accept_connection)

    print("Сервер запущен и ожидает подключений...")
    while True:
        events = selector.select(timeout=None)
        for key, _ in events:
            callback = key.data
            callback(key.fileobj)

start_server()
```

В приведенном примере кода демонстрируется, как можно создать мультиплексированный TCP-сервер, используя селекторы для асинхронной обработки множества соединений. Этот подход позволяет серверу одновременно обрабатывать запросы от нескольких клиентов без необходимости создавать отдельные потоки или процессы для каждого из них, значительно повышая производительность и эффективность использования ресурсов.

В начале кода мы импортируем необходимые модули: `selectors` для мультиплексирования ввода/вывода и `socket` для работы с сетевыми соединениями. Создается объект `selector` с помощью `selectors.DefaultSelector()`, который будет отвечать за отслеживание и обработку событий на сокетах.

Для обработки входящих соединений определяется функция `accept_connection`, которая автоматически вызывается, когда серверный сокет готов принять новое соединение. В этой функции с помощью метода `accept()` принимается входящее соединение, создается новый сокет для общения с клиентом, и этот новый сокет регистрируется в селекторе для отслеживания событий чтения данных. Это означает, что сервер будет уведомлен, когда появятся данные для чтения из этого сокета.

Функция `send_echo` задействуется, когда на одном из клиентских сокетов появляются данные для чтения. Она читает данные и отправляет их обратно клиенту, реализуя простую эхо-логику. Если клиент закрывает соединение, сокет удаляется из селектора и закрывается, освобождая системные ресурсы.

Основная логика сервера находится в функции `start_server`. Здесь создается и настраивается серверный сокет, который затем регистрируется в селекторе с указанием на интерес к событиям готовности к приему новых соединений. Серверный сокет настраивается как неблокирующий, что позволяет избежать остановки выполнения программы в ожидании событий на сокете.

После настройки сервер входит в бесконечный цикл, в котором активно используется `selector.select()` для ожидания событий на отслеживаемых сокетах. Как только происходит одно из этих событий, извлекается соответствующий обработчик, и вызывается с передачей ему сокета, с которым связано событие. Это позволяет эффективно распределить входящие запросы между функциями `accept_connection` и `send_echo`, обеспечивая асинхронную обработку соединений.

Таким образом, благодаря селекторам, сервер может одновременно обрабатывать множество соединений в одном потоке, что делает его более масштабируемым и производительным по сравнению с традиционными многопоточными или многопроцессорными подходами. Это демонстрирует мощь асинхронного программирования и позволяет разработчикам создавать высокоэффективные сетевые приложения.
